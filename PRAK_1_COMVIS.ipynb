```python
# install (aktifkan jika di colla)
# !pip install openvc-python matplotib
import cv2, numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("/content/lena.jpg") #sesuaikan path jika perlu
assert img is not None, "Gambar tidak ditemukan. Pastikan file sample.jpg ada di folder datasets."
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Tampilkan RGB vs Grayscale
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.title("RGB"); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(gray, cmap='gray'); plt.title('Grayscale'); plt.axis('off')
plt.show()
h, w = img.shape[:2]

#menampilkan tinggi dan lebar gambar

print("Height = {}, Width = {}".format(h, w))

(B, G, R) = img[100, 100]

# Menampilkan jumlah piksel
print("R={}, G={}, B={}".format(R, G, B))

# histogram RGB vs Grayscale
plt.figure(figsize=(10,4))
colors = ('b', 'r', 'g')
for i,c in enumerate(colors):
  hist = cv2.calcHist([img], [i], None, [256], [0, 256])
  plt.plot(hist, color=c, label=f'Channel {c}')
plt.legend(); plt.title("Histogram RGB"); plt.tight_layout(); plt.show()

plt.figure(); plt.hist(gray.ravel(), bins=256); plt.title("Histogram Grayscale"); plt.show()
```

```python
# hitung histogram (256 bins, range 0-255)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
# Plot Histogram
plt.plot(hist)
plt.title("Histogram Grayscale")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Jumlah Pixel")
plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

colors = ('r', 'g', 'b')
for i, col in enumerate(colors):
    hist = cv2.calcHist([img_rgb], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0, 256])
    plt.title('Histogram warna RGB')
    plt.xlabel('Intensitas Pixel')
    plt.ylabel('Jumlah Pixel')
```
